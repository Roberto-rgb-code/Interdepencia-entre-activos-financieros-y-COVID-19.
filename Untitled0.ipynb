```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Extraer datos de precios del BTC de CoinMarketCap
url = 'https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20130428&end=20230219'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Buscar la tabla de precios
table = soup.find('table', {'class': 'cmc-table'})

# Extraer los datos de la tabla
data = []
for row in table.find_all('tr'):
    cols = row.find_all('td')
    if len(cols) == 7:
        date = cols[0].text.strip()
        price = float(cols[4].text.replace(',', ''))
        data.append([date, price])

# Crear un DataFrame de pandas con los datos
df = pd.DataFrame(data, columns=['date', 'price'])
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')
```
